```python
# nall value tretment by droop
#  how to check nall value 
df_sales.isnull().sum()
```

```python
df_sales.dropna(subset=['city','delivery_status'],inplace=True)
```

```python
df_sales.columns
```

```python
mean_dlelivery_time=df_sales['delivery_time_mins'].mean()
print(mean_dlelivery_time)




```

```python
df_sales['delivery_time_mins'].fillna(mean_dlelivery_time,inplace=True)
```

```python
df_sales.isnull().sum()
```

```python
# check to duplicate value
df_sales.duplicated().sum()
```

```python
#remove duplicate record
df_sales.drop_duplicates(inplace=True)
```

```python
df_sales.duplicated().sum()
```

```python
#covert order data time to data time object
df_sales['order_date'] = pd.to_datetime(df_sales['order_date'])
```

```python
df_sales.columns
```

```python
# find min max and averge and total 
min_amount=df_sales['total_amount'].min()
max_amount=df_sales['total_amount'].max()
avrge_amount=df_sales['total_amount'].mean()
print(f"min total amount:______{min_amount:2f}")
print((f"max total amount:_____{max_amount:2f}"))
print((f"avrge_amount:_____{max_amount:2f}"))
```

```python
df_product.columns
```

```python
# top five product by total sales 
top_product=df_sales.groupby('product_id')['total_amount'].sum().nlargest(5)
print(top_product)
```

```python
# marg with product detail to get product name 
top_product_details=top_product.reset_index().merge(df_product,on= "product_id")
```

```python
print(top_product_details)
top_product_details.columns
```

```python
# total sales by city and sort and clean 

sales_by_city=df_sales.groupby( 'city')['total_amount'].sum().sort_values(ascending=False)
print(sales_by_city)
```

```python
# averge delivery time by city 
averge_delivery_time_by_city=df_sales.groupby( 'city')['delivery_time_mins'].mean().sort_values(ascending=False)
print(averge_delivery_time_by_city)
```

```python
#sales trend over time (e.g,month)
df_sales["month"]=df_sales['order_date'].dt.to_period("M")
print(df_sales["month"])
```

```python
monthly_sales=df_sales.groupby("month")["total_amount"].sum().sort_values(ascending=False)
print(monthly_sales)
```

```python
#sales by prodct catcory
sales_by_category=df_sales.merge(df_product,on='product_id')
```

```python
sales_by_category=sales_by_category.groupby( 'category')['total_amount'].sum().sort_values(ascending=False)
print(sales_by_category)
```
